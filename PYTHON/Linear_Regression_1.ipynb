```python
def clean_location_code(x):
    if x in ['suburban']:
        return 1
    elif x in ['rural']:
        return 0
    elif x in ['urban']:
        return 2    
    else:
        return x
```

```python
categorical['location_code'] = list(map(clean_location_code, categorical['location_code'])) 
categorical
```
